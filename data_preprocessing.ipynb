```python
import pandas as pd
import os
```

```python
def load_tsv_data(tsv_path: str) -> pd.DataFrame:
    """Load TSV data file with standard column names."""
    df = pd.read_csv(tsv_path, sep="\t", header=None, keep_default_na=False)
    df.columns = ["idx", "label", "sentence", "subject_text", "object_text", "subject_type", "object_type", "subject_group", "object_group", "label_choices"]
    return df

train_pth = "data/train.out"
test_pth = "data/test.out"
validation_pth = "data/test.out"
df_train = load_tsv_data(train_pth)
df_test = load_tsv_data(test_pth)
df_val = load_tsv_data(validation_pth)
```
