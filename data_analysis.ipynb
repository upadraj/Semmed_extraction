# Data Analysis for Biomedical Relation Extraction

This notebook analyzes the TSV dataset (train.out, test.out, validation.out) for fine-tuning.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

sns.set_style('whitegrid')
%matplotlib inline
```

## Load Data

```python
def load_tsv_data(tsv_path: str) -> pd.DataFrame:
    """Load TSV data file with standard column names."""
    df = pd.read_csv(tsv_path, sep="\t", header=None, keep_default_na=False)
    df.columns = ["idx", "label", "sentence", "subject_text", "object_text", 
                  "subject_type", "object_type", "subject_group", "object_group", "label_choices"]
    return df

df_train = load_tsv_data("../data/train.out")
df_test = load_tsv_data("../data/test.out")
df_val = load_tsv_data("../data/validation.out")

print(f"Train samples: {len(df_train)}")
print(f"Test samples: {len(df_test)}")
print(f"Validation samples: {len(df_val)}")
print(f"Total samples: {len(df_train) + len(df_test) + len(df_val)}")
```

## Dataset Overview

```python
print("\nTraining Data Sample:")
df_train.head()
```

```python
print("\nColumn Info:")
df_train.info()
```

## Relation Type Distribution

```python
# Count relation types
train_labels = df_train['label'].value_counts()
test_labels = df_test['label'].value_counts()
val_labels = df_val['label'].value_counts()

print("\nTop 20 Relation Types in Training Set:")
print(train_labels.head(20))
```

```python
# Plot distribution
plt.figure(figsize=(14, 6))
train_labels.head(20).plot(kind='bar')
plt.title('Top 20 Relation Types in Training Set')
plt.xlabel('Relation Type')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Entity Type Analysis

```python
print("\nTop Subject Types:")
print(df_train['subject_type'].value_counts().head(15))

print("\nTop Object Types:")
print(df_train['object_type'].value_counts().head(15))
```

```python
# Plot entity types
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

df_train['subject_type'].value_counts().head(15).plot(kind='barh', ax=axes[0])
axes[0].set_title('Top 15 Subject Types')
axes[0].set_xlabel('Count')

df_train['object_type'].value_counts().head(15).plot(kind='barh', ax=axes[1])
axes[1].set_title('Top 15 Object Types')
axes[1].set_xlabel('Count')

plt.tight_layout()
plt.show()
```

## Sentence Length Analysis

```python
df_train['sentence_length'] = df_train['sentence'].str.len()
df_train['word_count'] = df_train['sentence'].str.split().str.len()

print("\nSentence Statistics:")
print(df_train[['sentence_length', 'word_count']].describe())
```

```python
# Plot sentence length distribution
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].hist(df_train['sentence_length'], bins=50, edgecolor='black')
axes[0].set_title('Sentence Length Distribution (characters)')
axes[0].set_xlabel('Length')
axes[0].set_ylabel('Frequency')

axes[1].hist(df_train['word_count'], bins=50, edgecolor='black')
axes[1].set_title('Word Count Distribution')
axes[1].set_xlabel('Word Count')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
